**Python & Power BI**

```python
#Importamos librería
import pandas as pd
#Leeemos el archivo csv
df = pd.read_csv(r"/content/Dataset.csv",encoding="ISO-8859-1")
#Cambiamos texto en la columna sexo y borramos el total
df.loc[df.Sexo=="Homes","Sexo"] ="Hombres"
df.loc[df.Sexo=="Mulleres","Sexo"] ="Mujeres"
df.drop(df[df.Sexo=="Total"].index, inplace=True)
#Borramos columnas innecesarias
dfColumnaIn = df.drop(['CodTempo',"DatoN","Variable"], axis=1)
#Borramos números que estaban ubicados en los nombres de los espacios(Espazo)
df_final = dfColumnaIn.replace({'[1 2 3 4 5 6 7 8 9 0]':''},regex=True)
#Reiniciamos el índice
df_final.reset_index(inplace=True,drop=True)
#Resultado
df_final
```
